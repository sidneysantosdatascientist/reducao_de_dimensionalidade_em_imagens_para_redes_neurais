```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rgb_to_grayscale(image):
    """
    Converte uma imagem RGB para níveis de cinza usando a fórmula:
    gray = 0.299*R + 0.587*G + 0.114*B

    Args:
        image (numpy.ndarray): Imagem colorida no formato (altura, largura, 3).

    Returns:
        numpy.ndarray: Imagem em tons de cinza (altura, largura).
    """
    height, width, _ = image.shape
    grayscale_image = np.zeros((height, width), dtype=np.uint8)

    for i in range(height):
        for j in range(width):
            r, g, b = image[i, j]
            grayscale_image[i, j] = int(0.299 * r + 0.587 * g + 0.114 * b)

    return grayscale_image
```

```python
def binarize_image(grayscale_image, threshold=128):
    """
    Binariza uma imagem em tons de cinza com base em um limiar (threshold).

    Args:
        grayscale_image (numpy.ndarray): Imagem em tons de cinza (altura, largura).
        threshold (int): Valor de corte para binarização (0 a 255).

    Returns:
        numpy.ndarray: Imagem binarizada (altura, largura) com valores 0 ou 255.
    """
    height, width = grayscale_image.shape
    binary_image = np.zeros((height, width), dtype=np.uint8)

    for i in range(height):
        for j in range(width):
            binary_image[i, j] = 255 if grayscale_image[i, j] > threshold else 0

    return binary_image

# Exemplo de teste com uma imagem colorida fictícia (substitua pela sua imagem)
# Criando uma imagem RGB simples (3x3 para demonstração)
imagem_colorida = np.array([[[255, 0, 0], [0, 255, 0], [0, 0, 255]],
                            [[255, 255, 0], [0, 255, 255], [255, 0, 255]],
                            [[128, 128, 128], [64, 64, 64], [0, 0, 0]]], dtype=np.uint8)
```

```python
# Converte para níveis de cinza
gray_image = rgb_to_grayscale(imagem_colorida)
```

```python
# Converte para binarizado
binary_image = binarize_image(gray_image)
```

```python
# Exibindo os resultados
plt.figure(figsize=(10, 3))
```

```python
plt.subplot(1, 3, 2)
plt.title("Níveis de Cinza")
plt.imshow(gray_image, cmap="gray")
plt.axis("off")
```

```python
plt.subplot(1, 3, 3)
plt.title("Binarizada")
plt.imshow(binary_image, cmap="gray")
plt.axis("off")
```
